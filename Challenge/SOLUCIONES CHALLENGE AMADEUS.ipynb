```python
import pandas as pd

filename='bookings.csv.bz2'

pd.set_option('display.max_columns', None)
bc=pd.read_csv(filename, sep='^', usecols=['arr_port','pax','year'], chunksize=100000)

todocosa=pd.DataFrame()

for i, cosa in enumerate(bc):
    print(i)
    b=cosa.dropna()
    intermedio=b[b['year']==2013][['arr_port', 'pax']]
    aaa=intermedio.groupby('arr_port').sum()
    todocosa=todocosa.append(aaa)
    
resultado=todocosa['pax'].groupby('arr_port').sum().\
            sort_values(ascending=False).reset_index()
```
