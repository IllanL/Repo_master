# Basic Car Speed 

## Problem Statement

We would like to determine the speed of a car based on the bumpiness and slope of the road. For this, we have a list of the speed from a list of drivers.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("terrain.csv")
```

```python
X = df[["bumpiness","grade"]].values
y = df["target"]
```

```python
def draw(clf,X,y,h=0.05):
    
    plt.figure(figsize=(10,10))

    x_min, x_max = X[:, 0].min() - 0.05, X[:, 0].max() + .05
    y_min, y_max = X[:, 1].min() - .05, X[:, 1].max() + .05
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    
    if clf is not None:
        
        Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
        Z = Z.reshape(xx.shape)
        plt.scatter(xx,yy,c=Z)
    else:
        plt.scatter(xx,yy)
    
    plt.scatter(X[:,0],X[:,1],c=y,cmap="Paired")
```

```python
np.e
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/np.e, random_state=2)
```

```python
from sklearn.linear_model import LogisticRegression

clf=LogisticRegression()
Z= clf.fit(X_train,y_train)

draw(Z, X_test, y_test, 0.01)
```

```python
from sklearn.neighbors import KNeighborsRegressor
AAA = KNeighborsRegressor(n_neighbors=3)
Z= AAA.fit(X_train,y_train)

draw(Z, X_test, y_test, 0.01)
```

```python
Z.kneighbors_graph()
```

```python
from sklearn.metrics import precision_score, recall_score
from sklearn.metrics import confusion_matrix, classification_report

precision_score(y_test, Z.predict(X_test))
classification_report(y_test, Z.predict(X_test))

cross_val_score(Z,X,y,scoring="precission")
cross_val_score(Z,X,y,scoring="recall")
```

## DECISION TREE:

```python
from sklearn.tree import DecisionTreeClassifier

clfTree=DecisionTreeClassifier(min_samples_leaf=20, max_depth=3)

clfTree.fit(X_train, y_train)
```

```python
from sklearn.model_selection import GridSearchCV


cosa = GridSearchCV (DecisionTreeClassifier(), 
                     param_grid={"min_samples_leaf":[2,3,4,5,6], 
                                "max_depth":[2,3,4,5,6,7,8]}, 
                     
                     cv=5, scoring="accuracy")


cosa.fit(X,y)

print(cosa.best_params_)

print(cosa.best_score_)
```

```python
draw(cosa, X_test, y_test, 0.01)
```

```python
from sklearn.metrics import roc_curve, auc

target_pos=1

fp,tp,_ =roc_curve(y_test, cosa.best_estimator_.predict_proba(X_test)[:, target_pos])
```
