## PASAMOS A UNIR:

```python
import pandas as np

s=pd.read_csv('searches.csv', delimiter='^', chunksize=50000, nrows=700000)


resultados=pd.DataFrame([])

for cosa in s:
    
    #cosa=cosa.dropna()

    sfiltrado=cosa[cosa['Destination'].isin(['MAD','BCN','AGP'])].copy()

    sfiltrado['Date']=sfiltrado['Date'].apply(lambda x: str(x))
    sfiltrado['Date']=sfiltrado['Date'].apply(lambda x: x[0:7])

    sbueno=sfiltrado[['Date', 'Time', 'Destination']].copy()

    sa=sbueno.groupby(['Destination','Date'], as_index=False)['Time'].count()
  
    resultados=resultados.append(sa)
    
resultados=resultados.groupby(['Destination','Date'], as_index=False)['Time'].sum()
    
resultados.rename(columns={'Time':'Busquedas'}, inplace=True)

a=resultados.pivot('Date','Destination','Busquedas')

%pylab inline

a.plot()

a
```

# Y, FINALMENTE, APLICAMOS A TODO EL DATAFRAME:

```python
import pandas as np

s=pd.read_csv('searches.csv', delimiter='^', chunksize=50000)


resultados=pd.DataFrame([])

for cosa in s:
    
    #cosa=cosa.dropna()

    sfiltrado=cosa[cosa['Destination'].isin(['MAD','BCN','AGP'])].copy()

    sfiltrado['Date']=sfiltrado['Date'].apply(lambda x: str(x))
    sfiltrado['Date']=sfiltrado['Date'].apply(lambda x: x[0:7])

    sbueno=sfiltrado[['Date', 'Time', 'Destination']].copy()

    sa=sbueno.groupby(['Destination','Date'], as_index=False)['Time'].count()
  
    resultados=resultados.append(sa)
    
resultados=resultados.groupby(['Destination','Date'], as_index=False)['Time'].sum()
    
resultados.rename(columns={'Time':'Busquedas'}, inplace=True)

a=resultados.pivot('Date','Destination','Busquedas')

%pylab inline

a.plot()

a
```
